```python
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_csv('Annual SPEI Drought Index bux.csv')
df2 = pd.read_csv('Average Largest 1-Day Precipitation bux.csv')
df3 = pd.read_csv('Average Largest 5-Day Cumulative Precipitation bux.csv')
df4 = pd.read_csv('Precipitation amount during wettest days bux.csv')
df5 = pd.read_csv('Relative Humidity bux.csv')
df6 = pd.read_csv('Temperature-based Excess Mortality bux.csv')
df7 = pd.read_csv('bux-max-observed-timeseries-of-a (14).csv')
df8 = pd.read_csv('bux-mean-observed-timeseries-of-a (14).csv')
df9 = pd.read_csv('bux-min-observed-timeseries-of-a (14).csv')
df10 = pd.read_csv('bux-precipitation-observed-timeseries-of-a (14).csv')
```

```python
data1 = pd.concat([df7,df8,df9,df10], axis=1)
data1 = data1.loc[:, ~data1.columns.duplicated()]
data1
```

```python
data2 = pd.concat([df1,df2,df3,df4,df5, df6], axis=1)
data2 = data2.loc[:, ~data2.columns.duplicated()]
data2
```

```python
data2['Annual SPEI Drought Index (spei index)'] = data2['Annual SPEI Drought Index (spei index)'].fillna(data2['Annual SPEI Drought Index (spei index)'].mean())
```
