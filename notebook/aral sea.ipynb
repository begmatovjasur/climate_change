```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 1. Asosiy tarixiy ma'lumotlar
data_str = """0     10.17
1      9.74
2     10.67
3      9.82
4     10.21
5     10.13
6     10.19
7     10.01
8     10.42
9     10.19
10     9.93
11     9.97
12    10.80
13    10.58
14    10.13
15    10.81
16    11.81
17    11.02
18    11.49
19    10.91
20    11.20
21    11.38
22    11.95
23    11.95
24    13.02
25    13.40
26    14.57
27    15.44
28    14.97
29    15.09
30    16.80
31    17.70
32    18.80
33    20.30
34    21.90
35    22.90
39     22.9
40     23.9
41     25.0
42     28.0
43     30.0
44     32.0
45     34.0
46     35.0
47     36.0
48     37.0
49     42.0
50     43.5
51     49.8
52     50.6
53     55.8
54     58.6
55     70.0
56     78.7
57     86.3
58     90.0
59     92.1
60     95.3
61     97.6
62      102"""

# Matnli ma'lumotlarni ajratib olish
rows = [line.split() for line in data_str.split('\n')]
indices = [int(r[0]) for r in rows]
values = [float(r[1]) for r in rows]

# Tarixiy ma'lumotlar uchun DataFrame yaratish
df_historical = pd.DataFrame({
    'Indeks': indices, 
    'Yil': [1950 + i for i in indices],
    'Shorlik_gl': values,
    'Holat': 'Tarixiy' # Qaysi biri asl, qaysi biri prognoz ekanligini ajratish uchun
})

# 2. Modelni o'qitish (3-darajali Polinomial Regressiya)
X = df_historical[['Yil']]
y = df_historical['Shorlik_gl']

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# 3. 2013-yildan 2025-yilgacha prognoz qilish
future_years = np.arange(2013, 2026)
future_X_poly = poly.transform(future_years.reshape(-1, 1))
predictions = model.predict(future_X_poly)

# 4. Prognoz natijalarini yangi DataFrame'ga joylash
df_future = pd.DataFrame({
    'Indeks': future_years - 1950,
    'Yil': future_years,
    'Shorlik_gl': np.round(predictions, 2),
    'Holat': 'Prognoz'
})

# 5. Ikkala jadvalni (tarixiy va prognoz) pastma-past birlashtirish
df_combined = pd.concat([df_historical, df_future], ignore_index=True)

# Ma'lumotlarni yillar bo'yicha tartiblash (agar o'tkazib yuborilgan yillar o'rnini to'ldirish kerak bo'lsa)
df_combined = df_combined.sort_values(by='Yil').reset_index(drop=True)

# Natijaning oxirgi 15 ta qatorini ekranga chiqarish
print("Tayyor jadvalning oxirgi qismi:")
df_combined
```
