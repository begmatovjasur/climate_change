```python
import pandas as pd
```

```python
df1 = pd.read_csv('Annual SPEI Drought Index qash.csv')
df2 = pd.read_csv('Average Largest 1-Day Precipitation qash.csv')
df3 = pd.read_csv('Average Largest 5-Day Cumulative Precipitation qash.csv')
df4 = pd.read_csv('Precipitation amount during wettest days qash.csv')
df5 = pd.read_csv('qash-max-observed-timeseries-of-a (14).csv')
df6 = pd.read_csv('qash-mean-observed-timeseries-of-a (14).csv')
df7 = pd.read_csv('qash-min-observed-timeseries-of-a (14).csv')
df8 = pd.read_csv('qash-precipitation-observed-timeseries-of-a (13).csv')
df9 = pd.read_csv('Relative Humidity qash.csv')
df10 = pd.read_csv('Temperature-based Excess Mortality qash.csv')
```

```python
data1 = pd.concat([df5,df6,df7,df8], axis=1)
data1 = data1.loc[:, ~data1.columns.duplicated()]
data1.head()
```

```python
data2 = pd.concat([df1,df2,df3,df4,df9,df10], axis=1)
data2 = data2.loc[:, ~data2.columns.duplicated()]
data2
```

```python
data2['Annual SPEI Drought Index (spei index)'] = data2['Annual SPEI Drought Index (spei index)'].fillna(data2['Annual SPEI Drought Index (spei index)'].mean())
data2.isnull().sum()
```

```python
data1 = data1.rename(columns={'Category':'Year'})
data1.head(2)
```

```python
import plotly.express as px
parametr = [
    'Average Maximum Surface Air Temperature (Annual Mean °C)',
    '5-year smooth (Gaussian Smooth)',
    'Average Mean Surface Air Temperature (Annual Mean °C)',
    'Average Minimum Surface Air Temperature (Annual Mean °C)',
    'Precipitation (Annual Mean mm)'
]
for d in parametr:
    fig = px.line(data1, x='Year', y=d)
    fig.show()
```

```python
data2 = data2.rename(columns={'Category':'Year'})
data2.head(2)
```

```python
parametrs = [
    'Annual SPEI Drought Index (spei index)',
    'Average Largest 1-Day Precipitation (mm)',
    'Average Largest 5-Day Cumulative Precipitation (mm)',
    'Precipitation amount during wettest days (mm)',
    'Relative Humidity (%)',
    'Temperature-based Excess Mortality (factor)'
]
for d in parametrs:
    fig = px.line(data2, x='Year', y=d)
    fig.show()
```
