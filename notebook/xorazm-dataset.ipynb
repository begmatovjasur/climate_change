```python
import pandas as pd
```

```python
df1 = pd.read_csv('Annual SPEI Drought Index xor.csv')
df2 = pd.read_csv('Average Largest 1-Day Precipitation xor.csv')
df3 = pd.read_csv('Average Largest 5-Day Cumulative Precipitation xor.csv')
df4 = pd.read_csv('Precipitation amount during wettest days xor.csv')
df5 = pd.read_csv('Relative Humidity xor.csv')
df10 = pd.read_csv('Temperature-based Excess Mortality xor.csv')
df6 = pd.read_csv('xorazm-max-observed-timeseries-of-a (13).csv')
df7 = pd.read_csv('xorazm-mean-observed-timeseries-of-a (13).csv')
df8 = pd.read_csv('xorazm-min-observed-timeseries-of-a (13).csv')
df9 = pd.read_csv('xor-precipitation-observed-timeseries-of-a (13).csv')
```

```python
data1 = pd.concat([df6,df7,df8,df9], axis=1)
data1 = data1.loc[:, ~data1.columns.duplicated()]
data1
```

```python
data2 = pd.concat([df1,df2,df3,df4,df5,df10], axis=1)
data2 = data2.loc[:, ~data2.columns.duplicated()]
data2
```

```python
data2['Annual SPEI Drought Index (spei index)'] = data2['Annual SPEI Drought Index (spei index)'].fillna(data2['Annual SPEI Drought Index (spei index)'].mean())
data2.isnull().sum()
```
