```python
import pandas as pd
import plotly.express as px
```

```python
data0 = pd.read_csv(r'data\andijon\Annual SPEI Drought Index and.csv')
data2 = pd.read_csv(r'data\andijon\Average Largest 1-Day Precipitation-and.csv')
data3 = pd.read_csv(r'data\andijon\Average Largest 5-Day Cumulative Precipitation and.csv')
data4 = pd.read_csv(r'data\andijon\Precipitation amount during wettest days and.csv')
data5 = pd.read_csv(r'data\andijon\Relative Humidity and.csv')
data6 = pd.read_csv(r'data\andijon\Temperature-based Excess Mortality and.csv')
data7 = pd.read_csv(r'data\andijon\and-max-observed-timeseries-of-a (14).csv')
data8 = pd.read_csv(r'data\andijon\and-mean-observed-timeseries-of-a (14).csv')
data9 = pd.read_csv(r'data\andijon\and-min-observed-timeseries-of-a (14).csv')
data10 = pd.read_csv(r'data\andijon\and-precipitation-observed-timeseries-of-a (14).csv')
```

```python
df1 = pd.concat([data10,data9,data8,data7], axis=1)
df1 = df1.loc[:, ~df1.columns.duplicated()]
df1
```

```python
df1 = df1.rename(columns={'Category':'Year'})
```
